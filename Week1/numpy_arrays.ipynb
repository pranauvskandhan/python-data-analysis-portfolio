```python
# SLICING

# a way to create a sub-array based on the original array

a=np.array([0,1,2,3,4,5,6])
print(a[:3])
print(a[3:])
print(a[::2])
print(a[2 -1])
```

```python
b=np.random.rand(3,3)
print(b)

print(b[:2])

print(b[:2, 1:3])

# so in multi-dimensional arrays, the first argument is for selecting rows, and the 2nd argument is for selecting columns
```

```python
# slicing of an array is a view into the same data. this is called passing by reference. 
# so modifying the sub array will cosequently modify the original array

# here, we'll change the element as pos. [0,0] which is 0.485...., to 50. then we can see that the value in the original array is changed as well

sub_array=b[:2, 1:3]
print("sub array index [0,0] value before change", sub_array[0,0])
sub_array[0,0]=50
print("sub array index [0,0] value after change", sub_array[0,0])
print("original array index [0,1] value after change", b[0,1])
print(b)
```

# MANIPULATING TEXT W REGULAR EXPANSION

now we're gonna talk abt pattern matching in strings using regular expressions. Regular expressions, or regexes, are written in a condensed formatting language. in general, u can think of a regular expression as a pattern which u give to a regex processor w some source data.
the processor then parses that source data using that pattern, & returns chunks of text back to the data scientist or the programmer for further manipulation. There's really 3 main reasons u wanna do this - to check whether a pattern exists within some source data, to get all insurances of a complex pattern from some source data, or to clean ur source data using a pattern generally thru string slicing. Regexes arent trivial, but theyre a complex foundational techniques for data cleaning in data science applications, & a solid understanding of regexes willl help u quickly & efficiently manipulate text data for further data science application.


by the end of this, you'll understand the basics of regexes, how to define patterns for matching, how to apply these patterns to strings, & how to use the results of those patterns in data processing.

```python
# importing the regex lib

import re
```

```python
# match() checks for a match that is at the beginning of a string & returns a boolean
# search() checks for a match anywhere in the string & returns a boolean

txt="this is a good day"
if re.search("good", txt):
    print("wunderbar!")
else:
    print("alas :(")
```

```python
# in addition to checking for conditionals, we can segment a string. the work that regex does here is called tokenizing
# where the string is separated into substrings based on patterns. tokenizing is a core activity in natural language processing

txt="amy works diligently. amy gets good grades. our student amy is succesful"
re.split("amy", txt) # the split() fn will parse the string for us & return in chunks
```

```python
re.findall("amy", txt) # the findall() fn will parse the string for us & return in chunks
```

```python
re.search("amy", txt) # .search() looks for some pattern & returns a boolean
```

patterns & character classes

```python
grades="ACAABBAAAC"
print(grades.count("B")) # to count the no. of B
print(re.findall("[AB]", grades))
```

```python
re.findall("AC|AC", grades)
```

```python
# we can use the caret w the set operator to negate our results, eg., if we want to parse out only the grades which arent A's;
print(re.findall("[^A]", grades))
print(re.findall("^[^A]", grades))
```

Quantifiers

# quantifiers are the no. of times u want a pattern to be matched in order to match. the most basic quantifier is expresses as e{m,n} where e is the expression/character we're matching, m is the minimum no. of times u want it to b ematched & n is the maximum no.

```python
# lets using the grades as an example. how many times has this student been on a back-to-back A's streak?
re.findall("A{2,10}", grades)
```

```python
# we can also try n do this using single values and just repeating the pattern
re.findall("A{1,1}A{1,1}",grades)
```

```python
# if u have only 1 no. in the braces, its considered to be both m&n
re.findall("A{2}",grades)
```

```python
# using this we could find a decreasing trend in a student's grades
print(re.findall("A{1,1}B{1,1},C{1,1}",grades))
```
